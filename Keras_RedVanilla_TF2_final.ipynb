**Se importan librerias necesarias para el funcionamiento de nuestro MLP**

```
%tensorflow_version 2.x
```

```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

**Se cargan desde los datasets de keras y tensorflow el conjunto de datos fashion_mnist**

```
from keras.datasets import fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

**El preprocesamiento de datos se puede hacer más simple utilizando Keras utilizando to_categorical, un método de Keras que nos ayudará a convertir nuestras etiquetas, no sin antes, normalizar nuestros datos entre 0 y 1 (en los datos en X)**

```
from tensorflow.keras.utils import to_categorical

 
x_train = x_train.reshape((60000, 784))
x_train = x_train.astype('float32') / 255
y_train = to_categorical(y_train)
 
x_test = x_test.reshape((10000, 784))
x_test = x_test.astype('float32') / 255
y_test = to_categorical(y_test)
```

**Mezclamos nuestros conjuntos y armamos nuestro conjunto de validación**

```
import numpy as np
 
np.random.seed(42)
 
# barajamos el dataset
permutation = np.random.permutation(x_train.shape[0])
x_train = x_train[permutation]
y_train = y_train[permutation]
 
# obtenemos el conjunto de validación
num_val = 10000
x_val = x_train[:num_val]
x_train = x_train[num_val:]
y_val = y_train[:num_val]
y_train = y_train[num_val:]
```

# Arquitectura

** A diferencia de ternsorflow, podemos implementar un MLP mucho más fácil únicamente agregando capas densas en un modelo secuencial**

```
from keras import models, layers
 
model = models.Sequential()
model.add(layers.Dense(340, activation='relu', input_shape=(784,))) # 28x28=784
model.add(layers.Dense(200, activation='relu'))
model.add(layers.Dense(100, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
```

**Se compila y se entrena el modelo utilizando el optimizador de Adam**

```
model.compile(optimizer = 'Adam',
             loss = 'categorical_crossentropy',
             metrics = ['accuracy'])
```

```
train_log = model.fit(x_train, y_train, 
                      epochs=250, batch_size=228,
                     validation_data=(x_val, y_val))
```
